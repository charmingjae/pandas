```python
import pandas as pd
```

```python
# 엑셀 파일 불러오기
df = pd.read_excel(u'data/담당자별_판매량_통합.xlsx')
```

```python
print("----- Total Sales -----")
# 불러온 엑셀 파일 출력
df
```

```python
print(u'----- 3분기 250 이상 판매량 -----')
# 3분기 판매량이 250 이상인 행만 출력
df[(df[u'3분기']>=250)]
```

```python
df
```

```python
print(u'----- 분기별 판매량만 보기 -----')
# 분기별 판매량 출력
# 행은 전체 행, 열은 0, 3, 4, 5, 6 행만 출력
df.iloc[:, [0,3,4,5,6]]
```

```python
# 제품명이 핸드백인 데이터 프레임만 출력
product_name = u'핸드백'
# boolean indexing
handbag = df[(df[u'제품명']==product_name)]
handbag
```

```python
# 제품명인 핸드백인 데이터프레임에서 행 단위로 더한다. 여기서 컬럼 이름은 연간 판매량으로 지정
handbag_sum = pd.DataFrame(handbag.sum(axis=1), columns = [u'연간 판매량'])
```

```python
handbag_sum
```

```python
# handbag 데이터 프레임에 handbag_sum 데이터 프레임을 오른쪽에 조인
handbag_total = handbag.join(handbag_sum)
```

```python
handbag_total
```

```python
# handbag_total 데이터 프레임을 열 단위로 더해 '합계' 라는 이름의 컬럼으로 생성
# 기존의 handbag_total에 append를 하기 위해 전치를 시킨 후 아래에 삽입 시킨다.
handbag_sum2 = pd.DataFrame(handbag_total.sum(), columns=[u'합계'])
print(handbag_sum2)
handbag_total2 = handbag_total.append(handbag_sum2.T)
```

```python
handbag_total2
```

```python
# Sum을 하게 되면 숫자외에 문자열은 문자열 덧셈이 되기 때문에
# 임의로 변경을 시켜준다.
handbag_total2.loc[u'합계',u'제품명'] = product_name
handbag_total2.loc[u'합계',u'담당자'] = u'전체'
handbag_total2.loc[u'합계',u'지역'] = u'전체'
```

```python
handbag_total2
```

```python
print(u"---- 핸드백의 분기별 판매량  ----")
print(handbag)
print(u"---- 지역별 연간 판매량만 보기  ----")
print(handbag_total.sort_values(by=u'연간 판매량',ascending=True))
print(u"---- 핸드백 통합 판매량 보기  ----")
handbag_total2
```

```python
# 엑셀로 저장
excel_file_name = u'data/핸드백_통합_판매량.xlsx'
excel_total_file_writer = pd.ExcelWriter(excel_file_name, engine='xlsxwriter')
handbag_total2.to_excel(excel_total_file_writer, index=False, sheet_name=u'핸드백_통합_판매량')
excel_total_file_writer.save()
```
