```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
x = np.arange(-4, 5, 0.5)
y = 2 * x ** 2
print("y : ", y)
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
# Set Title
plt.title('y=x^2 Line Chart')
# Set xLabel
plt.xlabel('X-axis')
# Set yLabel
plt.ylabel('Y-axis')
# Set grid - 가운데 격자 선 표시 여부 설정
plt.grid(True)
plt.show()
```
