# Histogram & Pie Chart

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager
```

```python
# Fonts
font_location = '/Library/Fonts/Arial unicode.ttf'
font_name = font_manager.FontProperties(fname = font_location).get_name()
matplotlib.rc('font',family=font_name)
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
math = [76, 82, 84, 83, 90, 86, 85, 92, 72, 71, 100, 87, 81, 76, 94, 78, 81, 60, 79, 69, 74, 87, 82, 68, 79] 
plt.xlabel(u'시험 점수') 
plt.ylabel(u'도수(frequency)') 
plt.title(u'수학 시험의 히스토그램') 
plt.grid(True) 

# Histogram
plt.hist(math)
```

```python
fruit = [u'사과',u'바나나',u'딸기',u'오렌지',u'포도']
result = [7, 6, 3, 2, 2]
# tuple
explode_value = (0.1, 0, 0, 0, 0)
plt.figure()
# autopct = 비율 표시 형식
plt.pie(result, labels=fruit, autopct='%.1f%%', startangle=90, counterclock=False, explode=explode_value, shadow=True)
plt.show()
```
