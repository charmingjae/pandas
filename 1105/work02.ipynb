```python
from xml.etree.ElementTree import *
```

```python
tree = parse('./temp/Resource10/camp2.xml')
```

```python
# get root node
note = tree.getroot()
iterTag = note.find("body")
```

```python
# create list
camplist = list()
# item 태그 내 요소들을 이터레이팅
for parent in iterTag.iter("item"):
    camp = list()
    for item in parent:
        camp.append(item.text)
    camplist.append(camp)
```

```python
for i in range(len(camplist)):
    print("야영장 이름 : " , camplist[i][0])
    print("야영장 위도 : " , camplist[i][2])
    print("야영장 경도 : " , camplist[i][3])
    print("야영장 주소 : " , camplist[i][4])
    print("야영장 사이트 개수 : ", camplist[i][7])
    print("야영장 1일 숙박 수용인원 : " , camplist[i][10])
    print("야영장 편의 시설 : ", camplist[i][12])
    print()
```
