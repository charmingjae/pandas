# Scatter chart

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager
```

```python
# Font
font_location = '/Library/Fonts/Arial unicode.ttf'
font_name = font_manager.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
city = [u'서울',u'인천',u'대전',u'울산',u'부산',u'광주']
```

```python
# 위도(latitude)와 경도(longitude) 
lat  = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16]  
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85] 
```

```python
# 인구 밀도
pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995] 
size = np.array(pop_den) # 마커의 크기 지정
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y'] # 마커의 컬러 지정 
```

```python
# Scatter chart
plt.scatter(lon, lat, s=size, c=colors, alpha=0.5)
plt.xlabel(u'경도(logitude)')
plt.ylabel(u'위도(latitude)')
plt.title(u'지역별 인구 밀도(2017)')

for x, y, name in zip(lon, lat, city):
    plt.text(x, y, name)
plt.show()
```
