# maplotlib example 5

```python
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager
import numpy as np
```

```python
# Set Font
font_location = '/Library/Fonts/Arial Unicode.ttf'
font_name  = font_manager.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
x = np.arange(0, 5, 1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3
```

```python
# Set Line Properties
plt.plot(x, y1, '>--r', x, y2 , 's-g', x, y3 , ':db' , x, y4 , '-.Xc')
# Set 범례 문자열 및 위치
plt.legend(['data1','data2','data3','data4'], loc=4)
plt.title(u'그래프 형식 지정 예제')
plt.xlabel(u'0에서 4까지 1씩 증가하는 x축 값')
plt.ylabel(u'y축 값')
# ( 0 , 4 ) 좌표에 문자열 출력
plt.text(0, 4, u'문자열 출력')
plt.savefig('./temp/saveFigTest1.svg', dpi=100)
plt.show()
```
