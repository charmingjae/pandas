# matplotlib example 3

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from matplotlib import font_manager
```

```python
# x data
x = np.arange(-4.5 , 5, 0.5)

# y data
# y1 = 포물선
y1 = 2*x**2
# y2 = 직선
y2 = 5*x + 30
# y3 = 포물선
y3 = 4*x**2 + 30

plt.plot(x, y1)
plt.plot(x, y2)
plt.figure()
plt.plot(x, y3)
plt.show()
```
