# matplotlib example 1

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
x = np.arange(-4, 5, 0.5)
y = 2*x**2
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('title 입니다')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.grid(True)
plt.show()
```
