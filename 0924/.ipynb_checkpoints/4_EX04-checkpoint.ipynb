# Matplotlib Example 4

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10, 0.1)
```

```python
y1 = 0.3*(x-5)**2 + 1
y2 = -1.5*x + 3 
y3 = np.sin(x)**2
y4 = 10*np.exp(-x) + 1
```

```python
# 2 X 2 행렬로 이뤄진 하위 그래프에서 P에 따라 위치를 지정
# plt.subplot(m,n,p)
plt.subplot(2,2,1) # p는 1
plt.plot(x,y1)
plt.subplot(2,2,2)# p는 2
plt.plot(x,y2)
plt.subplot(2,2,3)# p는 3
plt.plot(x,y3)
plt.subplot(2,2,4)# p는 4
plt.plot(x,y4)

plt.show()
```
