```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager
```

```python
font_location = '/Library/fonts/Arial unicode.ttf'
font_name = font_manager.FontProperties(fname=font_location).get_name()
matplotlib.rcParams['font.family'] = 'Arial Unicode MS'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df1 = pd.read_csv('./temp/sea_rain.csv', encoding='utf8', sep='\t', index_col=u'년도')
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(121)
rain_plot = df1.plot(grid=True, style=['r--*','g-o','d:y','m-.p'], ax=ax1)
rain_plot.set_xlabel(u'연도')
rain_plot.set_ylabel(u'강수량')
rain_plot.set_title(u'연간 강수량')
rain_plot
```
