# Draw pandas graph

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager
```

```python
font_location = '/Library/Fonts/Arial unicode.ttf'
font_name = font_manager.FontProperties(fname=font_location).get_name()
matplotlib.rcParams['font.family'] = font_name
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
# Use Series
# style=['r--*','g-o','b:*','m-.p']
fig1 = plt.figure()
ax1 = fig1.add_subplot(121)
df_rain = pd.read_csv('./temp/sea_rain.csv', encoding='utf8', sep='\t', index_col=u'년도')
rain_plot = df_rain.plot(grid=True, style=['r--*','g-o','b:*','m-.p'], ax= ax1)
rain_plot.set_xlabel(u'엑스라벨')
rain_plot.set_ylabel(u'와이라벨')
rain_plot.set_title(u'타이틀')


# Use DataFrame
temperature = [25.2, 27.4, 22.9, 26.2, 29.5, 33.1, 30.4, 36.1, 34.4, 29.1] 
Ice_cream_sales = [236500, 357500, 203500, 365200, 446600, 574200, 453200, 675400, 598400, 463100] 
dict_data = {u'기온' : temperature, u'아이스크림 판매량' : Ice_cream_sales}

ax2 = fig1.add_subplot(122)
ice_cream_plot = pd.DataFrame(dict_data, columns=[u'기온', u'아이스크림 판매량'])
ice_cream_plot.plot.scatter(x=u'기온',y=u'아이스크림 판매량',grid=True, title=u'타이틀입니다',ax=ax2)
```

```python
fig2 = plt.figure()
ax3 = fig2.add_subplot(121)

grade_num = [5, 14, 12, 3] 
students = ['A', 'B', 'C', 'D'] 

df_grade = pd.DataFrame(grade_num, index=students, columns=['Students'])
grade_bar = df_grade.plot.bar(grid=True, ax=ax3)
grade_bar.set_xlabel(u'학점')
grade_bar.set_ylabel(u'학생수')
grade_bar.set_title(u'학점별 학생 수 막대 그래프')

math = [76,82,84,83,90,86,85,92,72,71,100,87,81,76,94,78,81,60,79,69,74,87,82,68,79]
ax4 = fig2.add_subplot(122)
df_hist = pd.DataFrame(math, columns=['Student'])
math_list = df_hist.plot.hist(bins=8, grid=True, ax=ax4)
math_list.set_xlabel(u'시험 점수')
math_list.set_ylabel(u'도수(frequency)')
math_list.set_title(u'수학 시험의 히스토그램')

fig3 = plt.figure()
ax5 = fig3.add_subplot(111)
fruit = [u'사과', u'바나나', u'딸기', u'오렌지', u'포도'] 
result = [7, 6, 3, 2, 2] 
df_fruit = pd.Series(result, index=fruit, name=u'선택한 학생수')
explode_value= (0.1, 0,0,0,0)
pie = df_fruit.plot.pie(figsize=(5,5), explode=explode_value,
                       counterclock=False, table=True,
                       shadow=True, autopct='%.1f%%',
                       startangle=90, ax=ax5
                    )
plt.show()
```
