```python
import json
import urllib
import urllib.request as urls
import pymysql
import ssl
```

```python
# 의존성 추가
context = ssl._create_unverified_context()
# ID & Secret Key
client_id = "클라이언트"
client_secret = "시크릿"

encText = urllib.parse.quote("python")

url = "https://openapi.naver.com/v1/search/book?query=" + encText+"&display=10&start=1"
request = urls.Request(url)
request.add_header("X-Naver-Client-Id", client_id)
request.add_header("X-Naver-Client-Secret", client_secret)
response = urls.urlopen(request, context=context)
rescode = response.getcode()
if(rescode == 200):
    response_body = response.read()
    json_data = json.loads(response_body.decode('utf-8'))
```

```python
json_data
```

```python
try:
    bookList =list()
    for items in json_data['items']:
        bookList.append(items)

    conn = pymysql.connect(host='localhost', user='root', password='비밀번호', db='bigdata', charset='utf8')
    curs = conn.cursor()
    for i in range(len(bookList)):
        sql = 'INSERT INTO book(b_title, b_author, b_publisher, b_pubcate, b_isbn, b_price) values(\'{0}\', \'{1}\', \'{2}\', \'{3}\', \'{4}\', {5})'.format(bookList[i]["title"], bookList[i]["author"], bookList[i]["publisher"], bookList[i]["pubdate"], bookList[i]["isbn"], int(bookList[i]["price"]))
        curs.execute(sql)
    conn.commit()
    conn.close()
except Exception as er:
    print(er)
```
